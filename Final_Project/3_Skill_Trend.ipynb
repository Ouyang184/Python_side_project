# How are in_demand skills trending for Data Analysts?

#### Methology
     1.Aggregate skill counts monthly
     2.Re-analyze based on percentage of total jobs
     3.Plot the monthly skill demand

```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import seaborn as sns
import ast

#Loading dataset
ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

# Changing job_posted_date to date time format
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_da_us = df[(df['job_country'] == 'United States') & (df['job_title'] == 'Data Analyst')].copy()
```

```python
df_da_us['job_posted_month_no'] = df_da_us['job_posted_date'].dt.month
```

```python
df_da_us_explode = df_da_us.explode('job_skills')
```

```python
df_da_us_pivot = df_da_us_explode.pivot_table(
    index='job_posted_month_no',
    columns= 'job_skills',
    aggfunc= 'size',
    fill_value= 0
)
df_da_us_pivot
```

```python
df_da_us_pivot.loc['Total'] = df_da_us_pivot.sum()

df_da_us_pivot = df_da_us_pivot[df_da_us_pivot.loc['Total'].sort_values(ascending= False).index]

df_da_us_pivot = df_da_us_pivot.drop('Total')

df_da_us_pivot
```

```python
da_total = df_da_us.groupby('job_posted_month_no').size()
```

```python
df_da_per = df_da_us_pivot.div(da_total /100, axis = 0)
```

```python
df_da_per = df_da_per.reset_index()

df_da_per['job_posted_month'] = df_da_per['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format = '%m').strftime('%b'))

df_da_per = df_da_per.set_index('job_posted_month')

df_da_per = df_da_per.drop(columns= 'job_posted_month_no')

df_da_per
```

```python
df_plot_trend = df_da_per.iloc[:, :5]

sns.lineplot(data = df_plot_trend, palette= 'tab10', legend= False, dashes= False)

sns.despine()
plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2022')
sns.set_theme(style='ticks')

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals= 0))

for i in range(5):
    plt.text(11.2,df_plot_trend.iloc[-1,i], df_plot_trend.columns[i])
```
