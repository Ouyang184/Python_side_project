```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np

ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
x = [1,2,3,4]
y = [2,3,4,5]
plt.plot(x,y)
plt.show
```

```python
df['job_posted_date'].head()
date_count = df['job_posted_date'].value_counts()
date_count = date_count.sort_index()
```

```python
date_count
```

```python
plt.plot(date_count.index, date_count)
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month

month_counts = df['job_posted_month'].value_counts()
month_counts = month_counts.sort_index()
month_counts
```

```python
plt.plot(month_counts.index, month_counts)
```
