```python
import pandas as pd
from datasets import load_dataset

ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_country'].value_counts().head(20)
```

```python
df['job_country'].isin(['China']).any()
```

```python
G_jobs =df[df['job_country'] == 'Germany']
```

```python
G_jobs =G_jobs[G_jobs['salary_year_avg'].notna()]
```

```python
G_jobs.groupby('job_title_short')[['salary_year_avg']].count()
```

```python
G_jobs.groupby('job_title_short')['salary_year_avg'].agg(['max', 'min','median']).sort_values(by = 'median', ascending= False)
```

```python
G_jobs.groupby(['job_title_short'])[['salary_year_avg', 'salary_hour_avg']].agg(
    {'salary_year_avg': ['max', 'min', 'median', 'count'],
     'salary_hour_avg': ['max', 'min', 'median', 'count']}).sort_values(by =[('salary_year_avg','max'),('salary_year_avg', 'count')] ,ascending= True)
```
